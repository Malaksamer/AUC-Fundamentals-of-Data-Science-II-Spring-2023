# Lists

- Lists can be thought of the most general version of a sequence in Python.

- Unlike strings, they are **mutable**, i.e. elements inside a list can be changed.

- Lists are constructed with brackets `[ ]` and commas `,` separating every element in the list.

# Creating a List

```python
weights = [65.0, 70.5, 72.3, 68.0, 77.2] # list of numbers
weights
```

```python
cities = ["London", "Paris", "New York", "Tokyo", "Berlin"] # list of strings
cities
```

```python
types = [1, 2.5, "hello", "world", 42, "python"] # list of different data types
types
```

# List of Lists

We can also create a list of lists. For example, combining the two list we just created, `cities` and `weights` into a new list `my_list`:

```python
my_list = [cities, weights]
my_list
```

```python
len(my_list)
```

# List Methods: `append`

The `append()` method adds an item to the end of the list

```python
print(cities)
len(cities)
```

```python
cities.append ("Aswan")
print(cities)
len(cities)
```

# List Methods: `pop`

The `pop()` method removes the item at the given index from the list and returns the removed item

```python
cities.pop() # pop (remove) the last element
```

```python
print(cities)
len(cities)
```

```python
cities.pop(1) # pop (remove) at the given index
```

```python
print(cities)
len(cities)
```

# List Methods: `reverse`

The `reverse()` method reverses the elements of the list

```python
print ("Before: ", cities)
```

```python
cities.reverse()
```

```python
print ("After: ", cities)
```

# List Methods: `sort`

The `sort()` method sorts the items of a list in ascending or descending order

```python
cities.sort()
cities
```

# List Methods: `index`

The `index()` method returns the index of the specified element in the list

```python
cities.index("Tokyo")
```

```python
# cities.index("Dubai")
```

# Dictionaries

Dictionaries in Python is a form of mapping, between keys and their corresponding value

# Constructing a Dictionary

```python
my_dict = {'key1':'value1','key2':'value2'} # Make a dictionary with {} and : to signify a key and a value
my_dict
```

```python
my_dict['key2'] # Call values by their key
```

```python
my_dict = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']} #dictionaries are very flexible
my_dict['key3']
```

# Dictionary Methods

```python
d = {'key1':1,'key2':2,'key3':3} # Create a typical dictionary
d
```

```python
d.keys() # Method to return a list of all keys 
```

```python
d.values() # Method to grab all values
```

```python
d.items() # Method to return tuples of all items  (we'll learn about tuples soon)
```
